Now we are going to calculate for K based on the convergence criteria of bessel function. Since in all 9part we need that, so that we will build a matrix initially with required r and lamda and will use that for iterative calculation of k 

```

def BesselMatrix(x,tolerance):#This one is always needed
  x=r
  lamda=np.linspace(0,10000,100000)
  lamda0=x[0]*lamda
  b=ss.jv(0,lamda0)
  for i in range(len(lamda0)-333):
    absol=np.absolute(b[i:i+333]) #Convergence condition
    suma=np.sum(absol)
    if suma<=tolerance:
      endarg=i
  lamdaN=lamda[0:i]
  M=len(x)
  N=len(lamdaN)
  Mat=np.zeros((M,N))
  for j1 in range(M):
    for j2 in range(N):
      Mat[j1,j2]=ss.jv(0,x[j1]*lamdaN[j2])
  

  

  return Mat,lamdaN,x
```

```
def Potential(r,tolerance,Kn,rho,Thick): #Con accounts for all other constants in the equation, either you can replace the Con by respective constants or can use the final value here
  Mat,lamdaN,x=BesselMatrix(r,tolerance)
  N=len(Thick)
  K=np.zeros((N,len(lamdaN)))
  K[N-1]=Kn
  for j1 in range(0,N-1):
    for j2 in range(len(lamdaN)):
      K[N-(j1+2),j2]=(K[N-(j1+1),j2]+((rho[N-(j1+2)]/rho[N-(j1+1)])*np.tanh(lamdaN[j2]*Thick[N-(j1+2)])))/((rho[N-(j1+2)]/rho[N-(j1+1)])+(K[N-(j1+1),j2]*np.tanh(lamdaN[j2]*Thick[N-(j1+2)])))
      print(K)
  c=np.multiply(Mat,K[0])
  (Row,Col)=c.shape
  Integ=np.zeros((Row,1))
  for i in range(Row):
    Integ[i,0]=si.simps(c[i],dx=0.1)
  return x,Integ,K
```

```
Kn=np.ones((1,len(lamdaN)))
d,I,K=Potential(5,0.01,Kn,rho,Thick)
print(d)
```
